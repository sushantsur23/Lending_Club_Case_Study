Methods to treat outliers:-

1) One way to deal with it could be to impute them with mean median or mode.

2) If the minimum is far away from the first quartile and the median. If we set the cleaning parameter to 0.6, Tukey's method detects Point A as an outlier and cleans it from the data set.

3) The Minkowski error is a loss index which does not deal and clean the outlier but it reduces the impact that outliers will have on the model.

4) We can use separate statistical models to treat significant number of outliers. The approach is to treat both groups (outlier and non-outlier) as two different groups and build individual model for both the groups and then combine their outputs

5) Transforming variables is also one of the ways to eliminate outliers. Natural log of a value reduces the variation caused by extreme values.
