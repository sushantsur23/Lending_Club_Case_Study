```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OrdinalEncoder

labelencoder = LabelEncoder()
df_rev['new_emp_title'] = labelencoder.fit_transform(df_rev['emp_title'])
# Viewing few rows of make and its encoded columns
df_rev[['emp_title','new_emp_title']].sample(20)
```

```python
import numpy as np
#Since I see that the NAn values are replaced with 28027 hence replacing the same 
df_rev['new_emp_title'] = df_rev['new_emp_title'].replace(to_replace=28027, value=np.nan)
```

```python
df_rev['new_emp_title']
```

Checking distribution before replacing null values

###### Imputing values for emp_title column

```python
import seaborn as sns
sns.distplot(df_rev['new_emp_title'])
```

```python
from sklearn.impute import SimpleImputer
# imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
# imp_mean = imp_mean.fit(df_rev[['new_emp_title']])
# emp_title_mean = imp_mean.transform(df_rev[['new_emp_title']])
# emp_title_mean

imp_mode = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imp_mode = imp_mode.fit(df_rev[['new_emp_title']])
emp_title_mode = imp_mode.transform(df_rev[['new_emp_title']])
```

```python
sns.distplot(emp_title_mode)
```

##### Conclusion :- 
Here we see that the distribution has not changed after imputing this method in place of mean hence the same concept we need to apply for other variables.

#### Imputing values for emp_length

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OrdinalEncoder

labelencoder = LabelEncoder()
df_rev['new_emp_length'] = labelencoder.fit_transform(df_rev['emp_length'])
# Viewing few rows of make and its encoded columns
df_rev[['emp_length','new_emp_length']].sample(1000)
```

```python
import numpy as np
#Since I see that the NAn values are replaced with 11 hence replacing the same 
df_rev['new_emp_length'] = df_rev['new_emp_length'].replace(to_replace=11, value=np.nan)
```

```python
import seaborn as sns
sns.distplot(df_rev['new_emp_length'])
```

```python
from sklearn.impute import SimpleImputer
# imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
# imp_mean = imp_mean.fit(df_rev[['new_emp_length']])
# emp_title_mean = imp_mean.transform(df_rev[['new_emp_length']])
# emp_title_mean

imp_mode = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imp_mode = imp_mode.fit(df_rev[['new_emp_length']])
emp_length_mode = imp_mode.transform(df_rev[['new_emp_length']])
```

```python
sns.distplot(emp_length_mode)
```

##### Conclusion

In this case also we can see that after imputing with the most frequent data the distribution doesn't change.
